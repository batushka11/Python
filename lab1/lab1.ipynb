```python
import numpy as np
import csv
import matplotlib.pyplot as plt

def get_column(filename, column_name):
    with open(filename, 'r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        header = next(reader)

        col_index = header.index(column_name)
        data = list(reader)
        column_data = np.array(data)[:,col_index]

        return column_data
    
```

```python
APC_numpy = get_column("russia_losses_equipment.csv", "APC").astype(int)

print(APC_numpy)
```

```python
daily_loses = np.diff(APC_numpy)

daily_loses = np.negative(daily_loses)

print(daily_loses)
```

```python
sort_loses = np.sort(daily_loses)[::-1]

print(sort_loses[:5])
```

```python
date_numpy = np.array(get_column("russia_losses_equipment.csv", "date"), dtype = "datetime64")

summer_loses = np.where((date_numpy >= np.datetime64('2023-06-01')) & (date_numpy <= np.datetime64('2023-08-31')))

apc_loses_summer = np.sum(daily_loses[summer_loses])

print(apc_loses_summer)
```

```python
apc_array_100_500 = daily_loses[-500:-99]

mean_loses = np.mean(apc_array_100_500)

print(mean_loses)
```

```python
plt.figure(figsize=(16,8), dpi=500)
plt.plot(date_numpy[0:200], daily_loses[0:200], linestyle='--', color='green', label='APC loses')
plt.legend()
plt.grid(color='grey', linestyle='--')
plt.xlabel("Date")
plt.ylabel("APC loses")
plt.title("200 last days APC loses")
plt.savefig('apc_loses.png', format='png')
plt.show()
```
